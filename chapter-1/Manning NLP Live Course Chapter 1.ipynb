```python
import pandas as pd
```

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize
```

```python
data_frame = pd.read_csv('/Users/prateekjain/work/data/Manning_Live_Project_NLP/stackexchange_812k.csv')
```

```python
data_frame.columns
```

```python
def to_lower_case(string):
    lower_case_string = string.lower()
    return lower_case_string
```

```python
def tokenize(string):
    tokenized_list = word_tokenize(string)
    return tokenized_list
```

```python
def collect_tokens(tokens):
    string = ','.join(tokens)
    return string
```

```python
data_frame['lower_case'] = data_frame['text'].apply(to_lower_case)
```

```python
data_frame['lower_case_split'] = data_frame['lower_case'].apply(tokenize)
```

```python
data_frame['lower_case_split_string'] = data_frame['lower_case_split'].apply(collect_tokens)
```

```python
data_frame.drop(['lower_case', 'lower_case_split'], axis=1)
```

```python
data_frame.to_csv("dataframe_solution.csv")
```
